```python
import pandas as pd

# D_S 01
print("Loading dataset 01:\n")
ds_01_path = pd.read_csv('ds_1/software_requirements_extended.csv')
print(ds_01_path.head(3))
print("\nshape:", ds_01_path.shape)
print("\ntypes:\n", ds_01_path['Type'].value_counts())
print("\nColumns:\n", list(ds_01_path.columns))
```

```python
import pandas as pd

# D_S 02
print("\nLoading dataset 02:\n")
ds_02_path = pd.read_csv('ds_2/Indeed_10k.csv', chunksize=1000)
print("path ready\n")
indeed_chunks = []
for chunk in ds_02_path:
    indeed_chunks.append(chunk)
indeed = pd.concat(indeed_chunks, ignore_index=True)
print("✅ Indeed dataset loaded!")
print("\nShape:\n", indeed.shape)
print("\nColumns:\n", list(indeed.columns))
print("\nFirst 3 rows:\n")
print(indeed.head(3))
```

```python
# See FR/NFR distribution (Topics task)
print("FR vs NFR in reference:", ds_01_path['Type'].value_counts())

# Extract job requirements (Kaggle task)  
job_reqs = indeed['Summary'].dropna()  # Your 10k requirements!
print("Job requirements ready:", len(job_reqs))
```

```python
# Reference distribution
print("Reference FR/NFR split:")
print(ds_01_path['Type'].value_counts())

# Job reqs ready for classification (Kaggle task)
print(f"\n10k+ job summaries ready: {len(job_reqs)} requirements")

# Preview: Length analysis (prioritization insight)
job_reqs['length'] = job_reqs.str.len()
print("\nAverage req length:", job_reqs['length'].mean())
print("Longest req:", job_reqs['length'].max(), "chars")
```

```python
# Simple FR/NFR classifier based on your reference dataset patterns
def classify_requirement(text):
    text_lower = text.lower()
    # FR patterns (what system DOES - from your 312 FR examples)
    fr_keywords = ['develop', 'build', 'create', 'implement', 'design', 'write', 'test']
    # NFR patterns (how well - from your 110 NFR examples)
    nfr_keywords = ['experience', 'years', 'degree', 'performance', 'secure', 'scalable']
    
    fr_score = sum(1 for word in fr_keywords if word in text_lower)
    nfr_score = sum(1 for word in nfr_keywords if word in text_lower)
    
    if fr_score > nfr_score:
        return 'FR'
    elif nfr_score > fr_score:
        return 'NFR'
    else:
        return 'Other'

# CLASSIFY ALL 10k job requirements
job_reqs_classified = job_reqs.apply(classify_requirement)
print("YOUR CLASSIFICATION RESULTS:")
print(job_reqs_classified.value_counts())
print("\nTop 3 FR examples:")
print(job_reqs[job_reqs_classified == 'FR'].head(3))
print("\nTop 3 NFR examples:") 
print(job_reqs[job_reqs_classified == 'NFR'].head(3))
```
